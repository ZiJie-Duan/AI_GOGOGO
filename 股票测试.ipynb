```python
import chardet

with open(r"C:\Users\lucyc\Desktop\AiStock\data\4_k_macd\zgcb.csv", 'rb') as file:
    rawdata = file.read()
encoding = chardet.detect(rawdata)['encoding']

print(encoding)
```

```python
date_d = []
open_d = []
high_d = []
low_d = []
close_d = []

with open(r"C:\Users\lucyc\Desktop\AiStock\data\4_k_macd\zgcb.csv", 'r', encoding=encoding) as file:
    for i in range(100):
        orig_data = file.readline()
        if i == 0:
            continue
        stock = orig_data.split(',')
        date_d.append(stock[0])
        open_d.append(float(stock[1]))
        high_d.append(float(stock[2]))
        low_d.append(float(stock[3]))
        close_d.append(float(stock[4]))
        print(stock)


```

```python
import plotly.graph_objects as go


fig = go.Figure(data=[go.Candlestick(x=date_d,
                open=open_d,
                high=high_d,
                low=low_d,
                close=close_d)])

fig.update_layout(title='股票价格',
                  xaxis_title='日期',
                  yaxis_title='价格',
                  xaxis_rangeslider_visible=False)

fig.show()
```
